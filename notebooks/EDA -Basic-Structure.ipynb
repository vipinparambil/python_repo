# EDA  - BASIC Structure

```python
import pandas as pd
import numpy as np
import os
```

## Import Data

```python
# set path for raw data
raw_data_path = os.path.join(os.path.pardir,'data','raw')
test_file_path = os.path.join(raw_data_path, "test.csv")
train_file_path = os.path.join(raw_data_path, 'train.csv')

train_file_path
```

### Read data

```python
train_df = pd.read_csv(train_file_path, index_col = "PassengerId")
test_df = pd.read_csv(test_file_path, index_col = "PassengerId")
```

```python
type(train_df)
```

### Basic Structure

```python
# Info functions

train_df.info()
```

```python
test_df.info()
```

```python
# Add missing column with 
test_df['Survived'] = -888
```

```python
test_df.info()
```

```python
# combine test and train data

df = pd.concat((train_df, test_df), axis=0, sort= True)
```

```python
## Filter
male_passengers = df.loc[df.Sex == 'male']
print("Male passengers count  : {0}".format(len(male_passengers)))
```

```python
## Male passengers with First class ticket

male_passsengers_with_first_class = df.loc[((df.Sex == 'male') & (df.Pclass == 1)),:]
print("Male passengers with first class count  : {0}".format(len(male_passsengers_with_first_class)))
```

## Summary statistics

```python
df.describe()
```

### Individual statistics

```python
#mean and median

print("Mean of Fare : {0}".format(df.Fare.mean()))
print("Median of Fare : {0}".format(df.Fare.median()))
```

```python
#Range
print("Range of Fare : {0}".format(df.Fare.max()- df.Fare.min()))

#25 percentile
print("25 Percentile of Fare : {0}".format(df.Fare.quantile(.25)))

#50 percentile
print("50 Percentile of Fare : {0}".format(df.Fare.quantile(.5)))

#25 percentile
print("75 Percentile of Fare : {0}".format(df.Fare.quantile(.75)))

#Varience
print("Varience of Fare : {0}".format(df.Fare.var()))

#SD
print("Varience of Fare : {0}".format(df.Fare.std()))
```
